Para construir la expresión en cada nivel del árbol, se elige una cifra o un operador diferente al de los nodos anteriores, y se van generando todas las posibles combinaciones. Si se llega a la cantidad deseada, se detiene la búsqueda y se retorna la solución encontrada.  

Para evitar repeticiones, en cada nivel del árbol se eliminan las opciones que ya fueron utilizadas en los nodos anteriores. Por ejemplo, si en el primer nivel se eligió la cifra 1, en el segundo nivel no se puede volver a elegir la cifra 1, ni tampoco el operador que se utilizó en el nivel anterior.

##Solución 1: Sin utilizar librerías, se crea dos funciones:

```
def permutaciones(iterable, r=None):
  '''
  Retorna permutaciones de elementos sucesivas de longitud r en el iterable
  '''
  conjunto = tuple(iterable)
  n = len(conjunto)
  r = n if r is None else r
  if r > n:
    return

  indices = list(range(n))
  ciclos = list(range(n, n-r, -1))
  yield tuple(conjunto[i] for i in indices[:r])

  while n:

    for i in reversed(range(r)): #permutación cíclica de una lista indices utilizando una lista auxiliar
      ciclos[i] -= 1             #ciclos que almacena la longitud de cada ciclo en la permutación.
      if ciclos[i] == 0:
        indices[i:] = indices[i+1:] + indices[i:i+1]
        ciclos[i] = n - i

      else:
        j = ciclos[i]
        indices[i], indices[-j] = indices[-j], indices[i]
        yield tuple(conjunto[i] for i in indices[:r])
        break

    else:
      return
```

```
import itertools

def encontrar_expresion2(numeros, n_objetivo):
    operadores = ['+', '-', '*', '/']
    for num_permutacion in itertools.permutations(numeros, 5): #Se genera todas las permutaciones posibles de longitud 5 de los elementos de numeros
                                                              #y las asigna a la variable num_permutacion en cada iteración del bucle. Máx 5 números en la expresión
        for op_permutacion in itertools.permutations(operadores, 4): #Se genera todas las permutaciones posibles de longitud 4 de los elementos de operaciones
                                                                     #y las asigna a la variable op_permutacion en cada iteración del bucle. 4 operaciones en la expresión

            expresion = f"{num_permutacion[0]}{op_permutacion[0]}{num_permutacion[1]}{op_permutacion[1]}{num_permutacion[2]}{op_permutacion[2]}{num_permutacion[3]}{op_permutacion[3]}{num_permutacion[4]}" #Se construye una expresión
                                                                                                                                       #aritmética en forma de cadena utilizando los operadores y números en las permutaciones actuales.

            if len(set(expresion)) == len(expresion): #Si la longitud del conjunto resultante es igual a la longitud de la cadena original expresion no había caracteres duplicados en la cadena original y por lo tanto la condición se
                                                      #cumple. Devuelve True si expresion no contiene caracteres duplicados y False en caso contrario.
                try:
                    if eval(expresion) == n_objetivo: #Se evalua la expresión y se compara con el número objetivo
                        print(expresion, "=", n_objetivo) #si el resultado es igual al número objetivo, se imprime la expresión y el número objetivo.
                except ZeroDivisionError: #Evita que se divida por cero en este caso se produce un error. La excepción se captura y se omite la expresión.
                    pass
```

```
import itertools

def encontrar_expresion3(numeros, n_objetivo):
    operadores = ['+', '-', '*', '/']
    n_list = []
    
    #Se crean las ramas del árbol
    for num_permutacion in itertools.permutations(numeros, 5): #Se genera todas las permutaciones posibles de longitud 5 de los elementos de numeros
                                                               #y las asigna a la variable num_permutacion en cada iteración del bucle. Máx 5 números en la expresión
        for op_permutacion in itertools.permutations(operadores, 4): #Se genera todas las permutaciones posibles de longitud 4 de los elementos de operaciones
                                                                     #y las asigna a la variable op_permutacion en cada iteración del bucle. 4 operaciones en la expresión

            expresion = f"{num_permutacion[0]}{op_permutacion[0]}{num_permutacion[1]}{op_permutacion[1]}{num_permutacion[2]}{op_permutacion[2]}{num_permutacion[3]}{op_permutacion[3]}{num_permutacion[4]}" #Se construye una expresión
                                                                                                                                       #aritmética en forma de cadena utilizando los operadores y números en las permutaciones actuales.

            # Se evita comparar las expresiones que contengan caracteres duplicados o que su resultado no sean números enteros enteros con el número objetivo.                                      
            if (len(set(expresion)) != len(expresion)) or (((round(eval(expresion),1)*10) % 10) != 0):
                continue 
            
            # Solo se compara las expresiones que tienen números enteros y no repiten carácteres con el número objetivo.
            try:                       
                if eval(expresion) == n_objetivo: #Se evalua las expresiones restantes y se compara con el número objetivo
                    print(expresion, "=", n_objetivo) #si el resultado es igual al número objetivo, se imprime la expresión y el número objetivo. 
                       
            except ZeroDivisionError: #Evita que se divida por cero en este caso se produce un error. La excepción se captura y se omite la expresión.
                pass 
```

## Solución 2: Algoritmo genético:

```
import random

#Defino las variables globales a utilizar en el algoritmo (mayúsculas)
NUMEROS_G = list(range(1, 10))
OPERADORES_G = ['+', '-', '*', '/']
N_OBJETIVO = 5
TAMANO_POBLACION = 10
NUMERO_GENERACIONES = 20 #Un algoritmo genético típicamente se itera de 
                         #50 a 500 o incluso más generaciones. 
```

```
def crear_expresion():
    '''
    Función para generar las expresiones con las restricciones del problema
    '''
   
    numeros = random.sample(NUMEROS_G,5) # Se cra una lista de 5 números aleatorios sin repetir
    #print(numeros)
    operadores = random.sample(OPERADORES_G,4) # Se crea una lista con los 4 operadores sin repetir
    #print(operadores)
    
    expresion =''.join([str(n) + op for n, op in zip(numeros, operadores)]) + str(numeros[-1]) # Se crea la expresión a partir de las dos listas anteriores
    #print(expresion)
                                                                                               # zip genera una lista de tuplas de número y signo
    return expresion
```

```
expresion = crear_expresion()
print(expresion)
```

```
def validacion_entero(expresion):
    '''
    Función para validar si la expresión evaluada corresponde un número entero
    '''
      
    if ((round(eval(expresion),1)*10) % 10) == 0: #En este caso el decimal debe ser cero
        return True
    else:
        return False 
```

```
validacion_entero(expresion)
```

```
def generar_poblacion_enteros(tamano):
    '''
    Función para generar población a partir de las expresiones que al evaluarsen den
    un número entero
    '''
    
    iterador = 0
    poblacion_ent = [] #inicializo la lista de población de expresiones de números enteros
    
    while iterador < len(range(tamano)):
        expresion = crear_expresion()
        
        if validacion_entero(expresion): #Mientras la expresion no corresponda a un número
                                         #entero se realizará el bucle
            poblacion_ent.append(expresion)
            iterador = iterador + 1
            
    return(poblacion_ent)
```

```
poblacion_ent = generar_poblacion_enteros(TAMANO_POBLACION)
print(poblacion_ent)
```

```
def evaluar_fitness(solucion):
    '''
    Función para evaluar a cada individuo de la población de expresiones que dan 
    números enteros y cuantificar cómo de bueno es cada individuo como solución
    al problema (fitness)
    '''
  
    try:
        validacion_sol = eval(solucion)
        fitness = 1/abs(N_OBJETIVO - validacion_sol + 1e-6) # Usamos un pequeño epsilon para evitar divisiones por cero
    
    except ZeroDivisionError: #Evita que se divida por cero en este caso se produce un error. 
        pass                  #La excepción se captura y se omite la expresión.
        
    return (fitness)
```

```
def selecionar_progenitores_ruleta(poblacion_ent):
    '''
    Se aplica la técnica conocida como "selección por ruleta", donde se asigna una
    probabilidad de selección a cada solución en función de su fitness. A medida
    que se avanza en el ciclo, se acumulan los valores de fitness de las soluciones
    y se compara con un valor aleatorio para seleccionar la solución correspondiente.
    '''
    
    fitness_total = 0
    
    for solucion in poblacion_ent: #Calcula el fitnees total de cada elemento de la poblacion
        fitness_p = [evaluar_fitness(solucion)]
        #print(fitness_p)
        fitness_total = sum(fitness_p) #Esta función toma una secuencia de valores numéricos y 
                                       #devuelve la suma total de todos los valores.
        
    #Se elige el primer progenitor
    
    r = random.uniform(0,fitness_total) #Genera un valor aleatorio entre 0 y el valor total. 
    #print(r)                                    #Este valor se usará para seleccionar el segundo padre.
    
    fitness_actual = 0 #Se inicializa variable para evaluar según el fitnnes o puntuación
    
    #Se realiza el primer ciclo para elegir el primer progenitor según su fitness
    for i, solucion in enumerate(poblacion_ent): #Comienza un ciclo a través de la población de soluciones. 
                                                 #El enumerate se utiliza para llevar un seguimiento del índice i y la solución solucion actual
        fitness_actual = fitness_actual + fitness_p[i]
        
        if fitness_actual >= r: #condicional para comparar con valor aleatorio
            progenitor1 = solucion
            break
        
        
    #Se elige el segundo progenitor
    
    r = random.uniform(0,fitness_total) #Genera un valor aleatorio entre 0 y el valor total. 
    #print(r)                                    #Este valor se usará para seleccionar el segundo padre.
    
    fitness_actual = 0 #Se inicializa variable para evaluar según el fitnnes o puntuación
    
    #Se realiza el primer ciclo para elegir el primer progenitor según su fitness
    for i, solucion in enumerate(poblacion_ent): #Comienza un ciclo a través de la población de soluciones. 
                                                 #El enumerate se utiliza para llevar un seguimiento del índice i y la solución solucion actual
        fitness_actual = fitness_actual + fitness_p[i]
        
        if fitness_actual >= r: #condicional para comparar con valor aleatorio
            progenitor2 = solucion
            break
        
    return(progenitor1,progenitor2)
```

```
progenitor1,progenitor2 = selecionar_progenitores_ruleta(poblacion_ent)
print(progenitor1)
print(progenitor2)
```

```
def cruces_progenitores(progenitor1,progenitor2):
    '''    
    Función que crea nuevos descendientes a partir de los dos progenitores dados utilizando una
    técnica de cruce llamada "cruce en un punto". En esta técnica, se elige un punto de corte 
    aleatorio en ambas cadenas de caracteres y se intercambian los fragmentos de cada progenitor 
    a partir del punto de corte.
    '''
    
    value=0
    while value==0:
        
        punto=random.randint(1,min(len(progenitor1),len(progenitor2))-1) # Se elige un punto de corte aleatorio 
        
        descendiente1=progenitor1[:punto]+progenitor2[punto:] #Se crea dos nuevos descendientes utilizando los 
        descendiente2=progenitor2[:punto]+progenitor1[punto:] #segmentos de cada progenitor que están en cada 
                                                              #lado del punto de corte. 
        
        if (len(descendiente1)==len(set(descendiente1))) or (len(descendiente2)==len(set(descendiente2))): #Se valida si
            value+=1                                        #los descendientes no tienen ni operadores ni números repetidos         
    
    return(descendiente1,descendiente2)
```

```
descendiente1,descendiente2 = cruces_progenitores(progenitor1,progenitor2)
print(descendiente1,descendiente2)
```

```
def mutacion(solucion):
    '''Función que muta a los descendientes, donde cada una de sus posiciones, puede modificarse
    con una probabilidad  p. Añade diversidad al proceso y evitar que el algoritmo caiga en 
    mínimos locales por que todos los individuos sean demasiado parecidos de una generación a otra.
    '''
    
    punto_mutacion= random.randint(0, len(solucion)- 1) #Se realiza en un punto de corte aleatorio dentro de la solución
                                                        #para introducir la mutación
   
    if punto_mutacion%2 != 0: #Si es impar, se mutará un operador. 
             
        tamano = len(list(set(OPERADORES_G) - set(solucion)))
        
        if tamano > 0: 
        
            #Se reemplaza el operador existente en ese lugar con otro operador aleatorio que
            # no esté presente en la solución original
            mutacion_op = random.choice(list(set(OPERADORES_G) - set(solucion)))
            nuevo_descendiente = solucion[:punto_mutacion] + mutacion_op + solucion[punto_mutacion+1:]
        
        else:
            #Si no hay operadores adicionales disponibles para la mutación, 
            #se selecciona un operador al azar de entre los operadores originales.
            mutacion_op = random.choice(OPERADORES_G)
            nuevo_descendiente = solucion[:punto_mutacion] + mutacion_op + solucion[punto_mutacion+1:]            
    
    else: #Si el índice del punto de corte es par, se mutará un número
        
        #Se reemplaza el número existente en ese lugar con un número aleatorio diferente
        # que no esté presente en la solución original. 
        mutacion_num = random.choice(list(set([str(i) for i in NUMEROS_G]) - set(solucion)))
        nuevo_descendiente = solucion[:punto_mutacion] + mutacion_num + solucion[punto_mutacion+1:]
   
    return nuevo_descendiente
```

```
def algoritmo_genetico():
    '''
    Función principal del algoritmo genetico
    '''
    
    poblacion = generar_poblacion_enteros(TAMANO_POBLACION) #Crear una población inicial 
                                                            #aleatoria de P individuos.
    
    for generacion in range(NUMERO_GENERACIONES): # Se evalua las generaciones
       
        progenitor1,progenitor2 = selecionar_progenitores_ruleta(poblacion) #Calcula la 
        #fortaleza (fitness) de cada individuo de la población y se selecciona los progenitores
        
        descendiente1,descendiente2 = cruces_progenitores(progenitor1,progenitor2) #Cruzar los 
        #progenitores seleccionados para generar un nuevo descendiente (crossover)
        
        descendiente1 = mutacion(descendiente1) # Se aplicar un proceso de mutación aleatorio
        descendiente2 = mutacion(descendiente2) # sobre los descendientes
        
        #Reemplazar la antigua población por la nueva.
        fitness_descendiente1 = evaluar_fitness(descendiente1) #Calculo fitness
        fitness_descendiente2 = evaluar_fitness(descendiente2) #Calculo fitness
        
        fitness_poblacion = [evaluar_fitness(solucion) for solucion in poblacion]
       
        #Para descendiente1
        fit_minimo = fitness_poblacion.index(min(fitness_poblacion)) #Encuentro el individuo con el fitness más bajo
        
        if fitness_descendiente1 > fitness_poblacion[fit_minimo]: #Reemplazo el individuo con menor fitness con el 
            poblacion[fit_minimo] = descendiente1                 #descendiente1
        
        #Para descendiente2    
        fit_minimo = fitness_poblacion.index(min(fitness_poblacion)) #Encuentro el individuo con el fitness más bajo
        
        if fitness_descendiente2 > fitness_poblacion[fit_minimo]: #Reemplazo el individuo con menor fitness con el 
            poblacion[fit_minimo] = descendiente2                 #descendiente2
            
        #Se encuentra la mejor solución para el número dado
        fit_mejor = max(poblacion,key=evaluar_fitness)
        
        if evaluar_fitness(fit_mejor) == 1:
            return(fit_mejor)
        
    return max(poblacion,key = evaluar_fitness)     
```

```
solucion = algoritmo_genetico()
print("La solución más óptima es ", solucion, "=", eval(solucion))
```
